```python
import pandas as pd
import numpy as np
import sys, os
```

```python
df_fake_1 = pd.read_csv('E:/Covid_Fake_News_Data/chosen/NewsFakeCOVID-19.csv')
nan_dropped_fake_1 = df_fake_1[df_fake_1['content'].notna()]
content_fake_1 = nan_dropped_fake_1['content']
content_fake_1  = pd.DataFrame(content_fake_1)

df_fake_2 = pd.read_csv('E:/Covid_Fake_News_Data/chosen/NewsFakeCOVID-19_5.csv')
nan_dropped_fake_2 = df_fake_2[df_fake_2['content'].notna()]
content_fake_2 = nan_dropped_fake_2['content']
content_fake_2  = pd.DataFrame(content_fake_2)

df_fake_3 = pd.read_csv('E:/Covid_Fake_News_Data/chosen/NewsFakeCOVID-19_7.csv')
nan_dropped_fake_3 = df_fake_3[df_fake_3['content'].notna()]
content_fake_3 = nan_dropped_fake_3['content']
content_fake_3  = pd.DataFrame(content_fake_3)

df_title_fake_1 = pd.read_csv('E:/Covid_Fake_News_Data/ClaimFakeCOVID-19_5.csv')
nan_title_dropped_fake_1 = df_title_fake_1[df_title_fake_1['title'].notna()]
titles_fake_1 = nan_title_dropped_fake_1['title']
titles_fake_1 = pd.DataFrame(titles_fake_1)
titles_fake_1.columns = ['content']
titles_fake_1

df_title_fake_2 = pd.read_csv('E:/Covid_Fake_News_Data/ClaimFakeCOVID-19_7.csv')
nan_title_dropped_fake_2 = df_title_fake_2[df_title_fake_2['title'].notna()]
titles_fake_2 = nan_title_dropped_fake_2['title']
titles_fake_2 = pd.DataFrame(titles_fake_2)
titles_fake_2.columns = ['content']
```

```python
combined_fake = [content_fake_1, content_fake_2, content_fake_3, titles_fake_1, titles_fake_2]
result_fake = pd.concat(combined_fake)
result_fake = pd.DataFrame(result_fake).reset_index(drop=True)
fake_labels = ['fake'] * len(result_fake)
fake_id = [i+1 for i in range(len(result_fake))]
result_fake.insert(0, "id", fake_id, True)
result_fake['label'] = fake_labels
result_fake.columns = ['id', 'content', 'label']
result_fake
```

```python
# Randomly sample 70% of your dataframe
df_fake_train = result_fake.sample(frac=0.7, random_state=42)
df_fake_train = df_fake_train.reset_index(drop=True)
df_fake_test = result_fake.loc[~result_fake.index.isin(df_fake_train.index)]
df_fake_test = df_fake_test.reset_index(drop=True)
```

```python
df_Real_1 = pd.read_csv('E:/Covid_Fake_News_Data/chosen/NewsRealCOVID-19.csv')
nan_dropped_Real_1 = df_Real_1[df_Real_1['content'].notna()]
content_Real_1 = nan_dropped_Real_1['content']

df_Real_2 = pd.read_csv('E:/Covid_Fake_News_Data/chosen/NewsRealCOVID-19_5.csv')
nan_dropped_Real_2 = df_Real_2[df_Real_2['content'].notna()]
content_Real_2 = nan_dropped_Real_2['content']

df_Real_3 = pd.read_csv('E:/Covid_Fake_News_Data/chosen/NewsRealCOVID-19_7.csv')
nan_dropped_Real_3 = df_Real_3[df_Real_3['content'].notna()]
content_Real_3 = nan_dropped_Real_3['content']
```
